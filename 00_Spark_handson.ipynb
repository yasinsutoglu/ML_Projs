# Hands-on-ML

```python
spark5 = SparkSession.builder.appName("Session_5").getOrCreate()
```

```python
spark5
```

```python
df = spark5.read.csv("Linear_regression_dataset.csv", header=True, inferSchema=True);df.show()
```

```python
df.count()
```

```python
len(df.columns)
```

```python
df.describe().show()
```

```python
df.printSchema()
```

```python
# output = b0 + b1 * var_1 + b2 * var_2 + b3 * var_3 + b4 * var_4 + b5 * var_5
```

```python
df.show(10, False) # first 10 items showed, unzipped data visualizaiton
```

```python
# vector assembling
from pyspark.ml.feature import VectorAssembler
```

```python
df_feat = VectorAssembler(inputCols=["var_1","var_2","var_3","var_4","var_5"],outputCol="features").transform(df)
```

```python
df_feat.printSchema()
```

```python
df_model = df_feat.select('features','output'); df_model.show(10,False)
```

```python
df_train, df_test = df_model.randomSplit([0.7, 0.3])
```

```python
df_train.count()
```

```python
df_test.count()
```

```python
# Linear Regression
from pyspark.ml.regression import LinearRegression
lin_reg = LinearRegression(featuresCol='features', labelCol='output')
```

```python
LR_model = lin_reg.fit(df_train)
```

```python
LR_model.intercept
```

```python
print(LR_model.coefficients) # [b1,b2,b3,b4,b5]
```

```python
test_preds = LR_model.evaluate(df_test)
```

```python
test_preds.predictions.show(10,False)
```

```python
test_preds.r2 # 0-1 arasındadır. 1e yaın olması daha iyidir. Modelin yordama kabiliyetini gösterir.
```

```python
test_preds.meanSquaredError # 0'a yakın olması iyidir. Hata orannın düşük olması demektir.
```

```python
test_preds.meanAbsoluteError
```
